```python
#1.  dataset -> titanic.csv, dan tampilkan
import pandas as pd 

dataset = pd.read_csv('titanic.csv')
dataset.head()  
```

```python
#2. test_dataset <- titanic_test.csv
import pandas as pd 

test_dataset = pd.read_csv('titanic_test.csv')
test_dataset.head()
```

```python
"""
3.train_data <- ambil dataset kolom fitur (Age, Fare). 
Hilangkan baris data yang terdapat missing values (catat posisi data yang hilang → pos_missing_train)
"""
import pandas as pd 

# Load dataset
data_train = pd.read_csv('titanic.csv')

# Ambil kolom 'Age' dan 'Fare'
train_data = data_train[['Age', 'Fare']]

# Catat posisi baris dengan missing values
pos_missing_train = train_data[train_data.isnull().any(axis=1)].index.tolist()

# Hapus baris dengan missing values
train_data = train_data.dropna()

# Posisi data yang hilang
print("Posisi data yang hilang:", pos_missing_train)

# Tampilkan sebagian data yang sudah dibersihkan
print(train_data.head())
```

```python
"""
4.test_data <- ambil test_dataset kolom fitur (Age, Fare). Hilangkan baris data yang 
terdapat missing values (catat posisi data yang hilang → pos_missing_test)
"""
import pandas as pd 

datatest = pd.read_csv('titanic_test.csv')

test_data = datatest[['Age', 'Fare']]
pos_missing_test = datatest[datatest.isnull().any(axis=1)].index.tolist()

# Hapus baris data dengan missing values di test dataset
test_data = test_data.dropna()
print("Posisi data yang hilang di test dataset", pos_missing_test)
test_data.head()
```

```python
"""
5. train_label <- ambil dataset kolom kelas (Survived), 
yang bukan pos_missing_train 
"""
print(pos_missing_train)
train_label = pd.DataFrame(dataset, columns=['Survived'])
train_label = train_label[~train_label.index.isin(pos_missing_train)]
train_label.head(8)
```

```python
# 6 test_label <- titanic_testlabel.csv, yang bukan pos_missing_test
import pandas as pd 

test_label = pd.DataFrame(pd.read_csv('titanic_testlabel.csv'), columns=['Survived'])
test_label = test_label[~test_label.index.isin(pos_missing_test)]
test_label.head(11)
```

```python
#7. Mencari nilai statistikal data train
train_min = train_data.min()
train_max = train_data.max()
print("Age")
print("Min = ",train_min[0])
print("Max = ",train_max[0])
print("\nFare")
print("Min = ",train_min[1])
print("Max = ",train_max[1])

#Normalisasi test data
train_data = (train_data-train_min)*(1-0)/(train_max-train_min)+0
train_data.head(8)
```

```python
#8. Normalisasi test data
test_data = (test_data-train_min)*(1-0)/(train_max-train_min)+0
test_data.head(8)
```

```python
from sklearn.neighbors import KNeighborsClassifier

class_result = {}
precision_ratio = {}
error_ratio = {}

for i in range(1, 11):
    print(f"kNN k={i}")
    kNN=KNeighborsClassifier(n_neighbors=i, weights='distance')
    kNN.fit(train_data, train_label)
    class_result[i] = kNN.predict(test_data)
    precision_ratio[i] = kNN.score(test_data, test_label)
    error_ratio[i] = 1-precision_ratio[i]
    print(class_result[i])
    print("\n")
```

```python
for i in range(1,11):
  print("Hasil klasifikasi untuk setiap nilai k", i)
  print('Akurasi = ', precision_ratio[i])
  print('error ratio = ', error_ratio[i])
  print('\n')


# Mendapatkan nilai maksimum
max_value = max(precision_ratio)


print("Nilai tertinggi:", precision_ratio)
```
