```python
"""
5. Lakukan pengisian missing value pada fitur Age dengan nilai mean dari
masing-masing class
"""
import pandas as pd 

dataset = pd.read_csv('titanic.csv')
print('Dataset Sebelum Pengisian\n', dataset)

# Mengisi nilai yang hilang pada kolom "Age" dengan rata-rata 
dataset['Age'] = dataset['Age'].fillna(dataset.groupby('Pclass')['Age'].transform('mean'))

print("\nDataset setelah pengisian missing value di kolom 'Age':\n", dataset)
```

```python
#6. Normalisasi dengan Min-Max(0-1)

import pandas as pd

# Memuat dataset (ganti dengan file dataset yang sesuai)
dataset = pd.read_csv('titanic.csv')
minval_age = dataset['Age'].min()
maxval_age = dataset['Age'].max()   
dataset['Age_MinMax'] = (dataset['Age'] - minval_age) / (maxval_age - minval_age)

minval_fare = dataset['Fare'].min()
maxval_fare = dataset['Fare'].max()
dataset['Fare_MinMax'] = (dataset['Fare'] - minval_fare) / (maxval_fare - minval_fare)

# Menampilkan hasil normalisasi
print("\nData Setelah Normalisasi Min-Max pada Kolom Age dan Fare (5 Baris Pertama):\n")
print(dataset[['Age', 'Fare', 'Age_MinMax', 'Fare_MinMax']].head())
```

```python
# 7. Normalisasi dengan Z-Core
import pandas as pd 

dataset = pd.read_csv('titanic.csv')
mean_age = dataset['Age'].mean()
std_age = dataset['Age'].std()
dataset['Age_ZScore'] = (dataset['Age'] - mean_age) / std_age

mean_fare = dataset['Fare'].mean()
std_fare = dataset['Fare'].std()
dataset['Fare_ZScore'] = (dataset['Fare'] - mean_fare) / std_fare

# Menampilkan hasil normalisasi Z-Score dalam format tabel (5 baris pertama)
print("\nData Setelah Normalisasi Z-Score pada Kolom Age dan Fare (5 Baris Pertama):\n")
zscore_data = dataset[['Age', 'Fare', 'Age_ZScore', 'Fare_ZScore']].head()

# Menampilkan data dalam bentuk tabel
print(zscore_data.to_string(index=False))
```

```python
# 8. Normalisasi Sigmoidal
import pandas as pd 
import numpy as np 


dataset = pd.read_csv('titanic.csv')
dataset['Age_Sigmoid'] = 1 / (1 + np.exp(-dataset['Age']))
dataset['Fare_Sigmoid'] = 1 / (1 + np.exp(-dataset['Fare']))

# Menampilkan hasil normalisasi Sigmoidal dalam format tabel (5 baris pertama)
print("\nData Setelah Normalisasi Sigmoidal pada Kolom Age dan Fare (5 Baris Pertama):\n")
sigmoid_data = dataset[['Age', 'Fare', 'Age_Sigmoid', 'Fare_Sigmoid']].head()

# Menampilkan data dalam bentuk tabel yang terstruktur
print(sigmoid_data.to_string(index=False))
```
