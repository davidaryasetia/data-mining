```python
"""
3. train_data <- ambil dataset kolom fitur (Sex, Age, Pclass, Fare). Lakukan pengisian
missing value pada fitur Age dengan nilai mean dari masing-masing class
"""
import pandas as pd 

dataset = pd.read_csv('titanic.csv')
train_data = dataset[['Sex', 'Age', 'Pclass', 'Fare']]

print("Dataset sebelum pengisian \n", dataset.head())

# Konversi kolom 'Sex' ke nilai numeric
train_data['Sex'] = train_data['Sex'].map({'male': 1, 'female': 0})
train_data['Age'] = train_data.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.mean()))

print("Data setelah pengisian missing value 'Age' : \n ", train_data.head())
```

```python
# 4. label <- ambil dataset kolom kelas (Survived)
import pandas as pd 

dataset = pd.read_csv('titanic.csv')
label = dataset['Survived']

# Tampil hasil 
print("Label (Survived): \n", label.head())
```

```python
"""
5. train_data <- lakukan normalisasi pada train_data dengan Min-Max 0-1 (catat nilai
min dan max setiap atribut)
"""
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Inisialisasi Min-Max Scaler
sc = MinMaxScaler(feature_range=(0, 1))

# 1. Metode Hold-out (70% train, 30% test)
xtrainHold, xtestHold, ytrainHold, ytestHold = train_test_split(train_data, label, test_size=0.3, random_state=42)

# Normalisasi Hold-out
train_dataHold = sc.fit_transform(xtrainHold)
test_dataHold = sc.transform(xtestHold)
print("Hasil normalisasi test_data Hold-out:\n", test_dataHold)

# 2. Metode K-Fold (k=10)
kf = KFold(n_splits=10, shuffle=True, random_state=42)

for train_index, test_index in kf.split(train_data):
    xtrainFold, xtestFold = train_data.iloc[train_index], train_data.iloc[test_index]
    ytrainFold, ytestFold = label.iloc[train_index], label.iloc[test_index]
    
    # Normalisasi K-Fold
    train_dataFold = sc.fit_transform(xtrainFold)
    test_dataFold = sc.transform(xtestFold)

    print("\nHasil normalisasi test_data K-Fold:\n", test_dataFold[:3], "...\n", test_dataFold[-3:])

# 3. Metode Leave-One-Out (LOO)
loo = LeaveOneOut()
accuracies = []

for train_index, test_index in loo.split(train_data):
    X_trainLoo, X_testLoo = train_data.iloc[train_index], train_data.iloc[test_index]
    y_trainLoo, y_testLoo = label.iloc[train_index], label.iloc[test_index]
    
    # Normalisasi LOO
    train_dataLoo = sc.fit_transform(X_trainLoo)
    test_dataLoo = sc.transform(X_testLoo)


print("\nHasil normalisasi test_data LOO:\n", test_dataLoo)
print("Akurasi rata-rata LOO: ", np.mean(accuracies))
```

```python
"""
7. Lakukan klasifikasi k-NN (k=3) untuk masing-masing pendekatan validasi dan
hitunglah error ratio-nya
"""
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np 
import pandas as pd 

error_ratio = pd.Series([None, None, None])
validasi = ['Hold-Out', 'K-Fold', 'LOO']

# ----- Metode 1: Hold-out -----
print(f"kNN k=3 validation Hold-Out")
knn = KNeighborsClassifier(n_neighbors=3)
xtrainHold, xtestHold, ytrainHold, ytestHold = train_test_split(train_data, label, test_size=0.3, random_state=42)
knn.fit(xtrainHold, ytrainHold)
class_result_Hold = knn.predict(xtestHold)
precision_ratio_Hold = accuracy_score(ytestHold, class_result_Hold)
error_ratio[0] = 1 - precision_ratio_Hold

# Tampilkan hasil prediksi dan error ratio
print(class_result_Hold)
print(f"Error ratio : {round(error_ratio[0] * 100, 2)}%")
print("\n")


# ----- Metode 2: K-Fold (k=10) -----
print(f"kNN k=3 validation K-Fold")
knn = KNeighborsClassifier(n_neighbors=3)

kf = KFold(n_splits=10, shuffle=True, random_state=42)
class_result_Fold_total = []

for train_index, test_index in kf.split(train_data):
    xtrainFold, xtestFold = train_data.iloc[train_index], train_data.iloc[test_index]
    ytrainFold, ytestFold = label.iloc[train_index], label.iloc[test_index]
    knn.fit(xtrainFold, ytrainFold)
    class_result_Fold = knn.predict(xtestFold)
    class_result_Fold_total.extend(class_result_Fold)
precision_ratio_Fold = accuracy_score(label, class_result_Fold_total)
error_ratio_Fold = 1 - precision_ratio_Fold
print(class_result_Fold_total)
print(f"Error ratio : {round(error_ratio_Fold * 100, 2)}%")
print("\n")

# ----- Metode 3: Leave-One-Out (LOO) -----
print(f"kNN k=3 validation LOO")
knn = KNeighborsClassifier(n_neighbors=3)

loo = LeaveOneOut()
class_result_Loo_total = []

for train_index, test_index in loo.split(train_data):
    xtrainLoo, xtestLoo = train_data.iloc[train_index], train_data.iloc[test_index]
    ytrainLoo, ytestLoo = label.iloc[train_index], label.iloc[test_index]
    knn.fit(xtrainLoo, ytrainLoo)
    class_result_Loo = knn.predict(xtestLoo)
    class_result_Loo_total.extend(class_result_Loo)

precision_ratio_Loo = accuracy_score(label, class_result_Loo_total)
error_ratio_Loo = 1 - precision_ratio_Loo

# Tampilkan hasil prediksi dan error ratio
print(class_result_Loo_total)
print(f"Error ratio : {round(error_ratio_Loo * 100, 2)}%")
print("\n")
```
