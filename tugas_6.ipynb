```python
# 1. dataset <- titanic.csv
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np 

dataset = pd.read_csv('titanic.csv')
dataset.head()
```

```python
#2. test_dataset <- titanic_test.csv
import pandas as pd 

test_dataset = pd.read_csv('titanic_test.csv')
test_dataset.head()
```

```python
"""3. train_data <- ambil dataset kolom fitur (Sex, Age, Pclass, Fare)
Lakukan pengisian missing value pada fitur Age dengan nilai mean dari
masing-masing class"""
train_data = dataset[['Sex', 'Age', 'Pclass', 'Fare']]


# Mengisi nilai yang hilang pada kolom "Age" dengan rata-rata 
train_data['Age'] = train_data['Age'].fillna(train_data.groupby('Pclass')['Age'].transform('mean'))
print("train_data setelah pengisian missing value di kolom 'Age':\n")
train_data['Age'].head()
```

```python
# 4. test_data <- ambil test_dataset kolom fitur (Sex, Age, Pclass, Fare).
test_data = test_dataset[['Sex', 'Age', 'Pclass', 'Fare']]
test_data.head()
```

```python
# 5. train_label <- ambil dataset kolom kelas (Survived)
train_label = pd.DataFrame(dataset, columns=['Survived'])
train_label.head()
```

```python
# 6. test_label <- titanic_testlabel.csv
import pandas as pd 

test_label = pd.read_csv('titanic_testlabel.csv')
test_label.head()
```

```python
# Mengonversi kolom 'Sex' menjadi numerik
train_data.loc[:, 'Sex'] = train_data['Sex'].map({'male': 1, 'female': 0})
test_data.loc[:, 'Sex'] = test_data['Sex'].map({'male': 1, 'female': 0})
```

```python
"""
7. Lakukan klasifikasi test_data terhadap train_data dengan Decision
Tree, dan berapakah error rationya?v
"""
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier


dtc=DecisionTreeClassifier()
dtc.fit(train_data, train_label)
acc=dtc.score(train_data, train_label)
err=round((1-acc)*100, 2)
print('\n\nError ratio = ', err, '%')


dot_data = tree.export_graphviz(dtc, out_file=None,
feature_names=train_data.columns.values)
```

```python
# 8. Visualisasi Decision Tree menggunakan Graphviz
import graphviz


dot_data = tree.export_graphviz(dtc, out_file=None, 
                                feature_names=['Sex', 'Age', 'Pclass', 'Fare'],
                                class_names=['Died', 'Survived'],
                                filled=True, rounded=True,
                                special_characters=True)


graph = graphviz.Source(dot_data)
graph.render("titanic_decision_tree",format="png")  # Simpan decision tree dalam bentuk file .pdf atau .png
graph.view()

graph = graphviz.Source(dot_data, format="png")
graph.render( view=True)
```
